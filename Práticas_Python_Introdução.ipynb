```python
altura = 1.79
```

```python
if (altura < 2):
  print("Você não pode jogar de Pivô no Basquete")
elif (altura > 1.50) and (altura < 1.90):
  print("Você pode jogar de Armador ou Ala")
else:
  print("Você pode ser um Pivô")  
```
